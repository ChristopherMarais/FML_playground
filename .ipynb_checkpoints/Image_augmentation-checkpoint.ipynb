```python
# function to flip images
def flip_image_ar(image_array):
    flipped_lst = []
    for idx in range(0,image_array.shape[3]):
        image_ob = Image.fromarray(image_array[:,:,:,idx], mode='RGB')
        
        lr_image = image_ob.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
        # tb_image = image_ob.transpose(Image.Transpose.FLIP_TOP_BOTTOM)
        
        lr_image_ar = np.asarray(lr_image)
        tb_image_ar = np.asarray(tb_image)
        original_image_lst = [lr_image_ar, tb_image_ar]
        flipped_lst.append(original_image_lst)
    flipped_ar = np.array(flipped_lst)
    flipped_ar = np.moveaxis(flipped_ar, source=[0, 1], destination=[4, 3])
    return flipped_ar


# function to rotate images
def rotate_image_ar(image_array, images_count=10):
    grad_split = 360/images_count
    rotated_lst = []
    for idx in range(0,image_array.shape[3]):
        image_ob = Image.fromarray(image_array[:,:,:,idx], mode='RGB')
        original_image_lst = []
        
        for grad in range(grad_split,360, grad_split):
            rotated_image = image_ob.rotate(grad)
            rotated_image_ar = np.asarray(rotated_image)
            
            original_image_lst.append(rotated_image_ar)
        rotated_lst.append(original_image_lst)
    rotated_ar = np.array(rotated_lst)
    rotated_ar = np.moveaxis(rotated_ar, source=[0, 1], destination=[4, 3])
    return rotated_ar

# function to change brightness of images
def brighten_image_ar(image_array, images_count = 10):
    bright_values = np.linspace(0.1, 2, images_count)
    bright_lst = []
    for idx in range(0,image_array.shape[3]):
        image_ob = Image.fromarray(image_array[:,:,:,idx], mode='RGB')
        original_image_lst = []
        
        for brightness in bright_values:
            bright_image = ImageEnhance.Brightness(image_ob).enhance(brightness)
            bright_image_ar = np.asarray(bright_image)
        
            original_image_lst.append(bright_image_ar)
        bright_lst.append(original_image_lst)
    bright_ar = np.array(bright_lst)
    bright_ar = np.moveaxis(bright_ar, source=[0, 1], destination=[4, 3])
    return bright_ar

```
